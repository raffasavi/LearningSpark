```python
def check_df(dataframe):
    pd.set_option('display.max_rows', None)  # Mostra tutte le righe
    pd.set_option('display.max_columns', None)  # Mostra tutte le colonne
    pd.set_option('display.width', 1000)  # Allarga l'output per evitare troncamenti
    pd.set_option('display.max_colwidth', None)  # Evita che il contenuto delle colonne venga troncato

    print(f"""###############################################################\n######################### ANALISI DF ##########################""")
    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    display(dataframe.head(3))  # Usa display() invece di print()
    print("##################### Tail #####################")
    display(dataframe.tail(3))  # Usa display() invece di print()
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("\n\n")

check_df(df1)

```
